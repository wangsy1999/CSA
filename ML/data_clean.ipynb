```python
import numpy as np
import pandas as pd
```

```python
def date_transfer(date):
    date2 = ""
    date2 += date[6:10]
    date2 += "-"
    date2 += date[0:2]
    date2 += "-"
    date2 += date[3:5]
    return date2
print(date_transfer("07/25/2017 08:00:00 AM"))
```

```python
AP1 = pd.read_csv("pm2021.csv")
AP2 = pd.read_csv("pm2022.csv")
AP = AP1.append(AP2)
AP["Date"] = AP["Date"].apply(date_transfer)
AP = AP.drop(AP[AP['STATE_CODE']!=36].index)
AP = AP[['Date','Daily Mean PM2.5 Concentration','DAILY_AQI_VALUE']]
AP
APD = AP["Date"].unique()
# temp = data[data["DATE"] == d2[0]]
# temp_column = temp[column_list[4]]
# print(temp_column)
# temp_column = temp_column.dropna(axis=0,how='any')
# print(temp_column.unique())
AP_column_list = ['Daily Mean PM2.5 Concentration','DAILY_AQI_VALUE']
a = []
for i in APD:
    temp = AP[AP["Date"] == i]
    b=[i]
    for tcolumn in AP_column_list:
        temp_column = temp[tcolumn]
        b.append(np.mean(temp_column))
    a.append(b)
AP_clean = pd.DataFrame(a)
AP_clean
```

```python
BCC = pd.read_csv("Bicycle_Counts.csv")
BCC["date"] = BCC["date"].apply(date_transfer)
BCC

BCCD = BCC["date"].unique()
# temp = data[data["DATE"] == d2[0]]
# temp_column = temp[column_list[4]]
# print(temp_column)
# temp_column = temp_column.dropna(axis=0,how='any')
# print(temp_column.unique())

a = []
for i in BCCD:
    temp = BCC[BCC["date"] == i]
    b=[i]
    temp_column = temp["counts"]
    b.append(np.sum(temp_column))
    a.append(b)
BCC_clean = pd.DataFrame(a)
BCC_clean
```

```python
def drop_col(df, cutoff=0.5): # 如果这一列中有90%以上的缺失值，那么就从df中删除这一列
    n = len(df)
    cnt = df.count() # 对列进行非nan值计数
    cnt = cnt / n # 求出非nan值的百分比
    return df.loc[:, cnt[cnt >= cutoff].index] 
test = drop_col(test)
test
test = test.dropna(axis=1,how='any')
test
```
